<h2>1. Chuẩn hóa dữ liệu</h2>

```python
import pandas as pd
import numpy as np
df = pd.read_csv("./google-play-store.csv")
df.head()
```

```python
df.info()
```

Hiệu chỉnh các dữ liệu rating bị thiếu:

```python
df[np.isnan(df["Rating"])]
```

```python
meanRating = df["Rating"].mean()
df.loc[np.isnan(df["Rating"]), "Rating"] = meanRating
```

Chuẩn hóa dữ liệu Size:

```python
df["Size"] = df["Size"].replace({'M':''}, regex=True)
for index, row in df.iterrows():
    if row["Size"] == "Varies with device":
        df.loc[index, "Size"] = 0.0
    elif row["Size"][-1] == "k":
        df.loc[index, "Size"] = float(row["Size"][:-1])/1000
```

```python
df["Size"] = pd.to_numeric(df["Size"], errors='coerce')
meanSize = df["Size"].mean()
df.loc[(df["Size"] == 0.0) | (np.isnan(df["Size"])), "Size"] = meanSize
```

Hiệu chỉnh số lượng Installs:

```python
df["Installs"] = df["Installs"].replace({'\+':''}, regex=True)
df["Installs"] = df["Installs"].replace({',':''}, regex=True)
df["Installs"] = pd.to_numeric(df["Installs"], errors='coerce').fillna(0.0).astype(int)
```

Hiệu chỉnh Type & Price:

```python
df[(df["Type"] != "Free") & (df["Type"] != "Paid")]
```

```python
df.drop([10472])
```

```python
df.loc[(df["Type"] != "Free") & (df["Type"] != "Paid"), "Type"] = "Free"
df.loc[df["Price"] == "Everyone", "Price"] = 0
df["Price"] = df["Price"].replace({'\$':''}, regex=True)
df["Price"] = pd.to_numeric(df["Price"], errors='coerce')
```

```python
df.info()
```

Save csv:

```python
df.to_csv('./handled-google-play-store.csv')
```
